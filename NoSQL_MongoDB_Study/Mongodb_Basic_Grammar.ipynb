## [ 2. SQL과 간단 비교 - mongodb collection 생성/변경 ]
* collection 생성 (원하는 타임으로 데이터를 바로 넣으면 됨)
  - PRIMARY KEY를 위한 별도 컬럼 만들 필요 없음. 
  - mongodb는 collection에서 _id가 각 Document마다 자동생성되어 primary key 역햘을 함
  - 컬럼마다 데이터 타입을 정할 필요 없음 ("컬럼명": 컬럼값 이 기본 형태임)
<br>
<img src="https://www.fun-coding.org/00_Images/mongodb_create.png" /> 
<br>
<br>
* collection 구조 변경 (기존 Document에 컬럼 추가/삭제 필요없을 시는 새로운 Document에만 필요한 컬럼을 추가 또는 삭제해서 넣으면됨)
  - ALTER TABLE은 기본적으로 collection에서는 필요 없음 
  - 일부 기존 Document에도 컬럼과 컬럼값을 넣거나 삭제해야 한다면 다음과 같은 형태로는 가능함
<br>
<br>
<br>
* 기존 Document에도 컬럼과 컬럼값 추가시 
  - SQL: ALTER TABLE people ADD COLUMN join_date DATETIME
  - mongodb: db.people.updateMany({ }, { $set: { join_date: new Date() } })

<br>

* 기존 Document에도 컬럼과 컬럼값 삭제시 
  - SQL: ALTER TABLE people DROP COLUMN join_date
  - mongodb: db.people.updateMany({ }, { $unset: { "join_date": "" } })

### Document 읽기(검색)
-----

#### (1) Document 읽기(검색) - findOne, find
- findOne : 매칭되는 <b>한개</b>의 document 검색
- find : 매칭되는 <b>list</b> of document 검색

<br>

#### (2) Document 읽기(검색) 문법
<img src="https://www.fun-coding.org/00_Images/mongodb_find_structure.png" /> 

<br>

- <b>Collection에 있는 모든 데이터를 가져오는 경우</b>
> db.[collection name].find()

- <b>Collection에 있는 모든 데이터중에 특정 key와 Value만 가져오는 경우</b>
> db.[collection name].find({ }, { key_name1 : 1, key_name2 : 1 })  
> <br>
> <b>{ }</b><br>
> : 모든 데이터를 의미, 이 괄호안에 특정 전제조건을 붙여 사용할 수 있다
> <br>
>
> <b>{ key_name1 : 1, key_name2 : 1 }</b><br>
> : 1 은 key_name1 과 key_name2의 정보만 가져올 것이라는 의미한다.
> <br>
>
> Total. 모든 데이터의 key_name1과 keynam2의 정보만을 가져오고 그 외의 것들은 가져오지 않는다.  
> cf. 이때, _id는 default 값으로 항상 출력된다.  
> 만약, _id 출력을 원치 않는다면, <b>_id : 0</b>옵션을 함께 명시한다. 

- <b>해당 Key와 Value 조건을 만족하는 데이터만 출력을 원할 경우</b>
> db.[collection name].({ key : value })

- <b>조건의 AND</b>
> db.[collcetion name].({ key1 : value1, key2 : value2 })

- <b>조건의 OR</b>
> db.[collection name].( { $or: [ { key1 : value1, key2 : value2 } ] } )

#### (3) 비교 문법
|Operation|Symbol|Description|
|:-----:|:-:|:-------------------------------------------------------:|
|`$eq`|=|Matches values that are <b><U>eq</U></b>ual to specified value.|
|`$gt`|>|Matches values that are <b><U>g</U></b>reater <b><U>t</U></b>han a specified value.|
|`$gte`|>=|Matches values that are <b><U>g</U></b>reater <b><U>t</U></b>han or <b><U>e</U></b>qual to a specified value.|
|`$in`| |Matches any of the values specified in an array.<br>ex) {age : { $in : [5, 15] } } : age의 값이 5또는 15인 것|

|Operation|Symbol|Description|
|:-----:|:-:|:-------------------------------------------------------:|
|`$lt`|<|Matches values that are <b><U>l</U></b>ess <b><U>t</U></b>han a specified value.|
|`$lte`|<=|Matches values that are <b><U>l</U></b>ess <b><U>t</U></b>han or <b><U>e</U></b>qual to a specified value.|
|`$ne`|!=|Matches all values that are <b><U>n</U></b>ot <b><U>e</U></b>qual to a specified value.|
|`$nin`| |Matches none of the values specified in an array.<br>ex) {age : { $nin : [5, 15] } } : age의 값이 5도 15도 아닌 것|

<br>

#### - 비교 문법 예제

- comparison operator

```python
db.people.find({ age: { $gt: 25 } })  
db.people.find({ age: { $lt: 25 } })  
db.people.find({ age: { $gt: 25, $lte: 50 } })
db.people.find( { age: { $nin: [ 5, 15 ] } } ))
```

- `/검색키워드/` 를 이용하여 특정 글자를 내포하는 데이터를 검색할 수 있다.

```python
db.people.find( { user_id: /bc/ } )  
db.people.find( { user_id: { $regex: /bc/ } } )  
                                                
db.people.find( { user_id: /^bc/ } )
db.people.find( { user_id: { $regex: /^bc/ } } )  
```

- 보류

```python
db.people.find( { status: "A" } ).sort( { user_id: 1 } )   
db.people.find( { status: "A" } ).sort( { user_id: -1 } )   
db.people.count()  
db.people.find().count()  
                                                
db.people.count( { user_id: { $exists: true } } )  
db.people.find( { user_id: { $exists: true } } ).count()  
                                                
db.people.count( { age: { $gt: 30 } } )  
db.people.find( { age: { $gt: 30 } } ).count()  
                                               
db.people.distinct( "status" )   
db.people.findOne()  
db.people.find().limit(1)
```

### Document 수정
-----


<img src="https://www.fun-coding.org/00_Images/mongodb_update_structure.png" />


- `$set` : field 값을 설정(수정)

- `$inc` : field 값을 증가시키거나, 감소시킴
    - ex. { $inc : { age : 2 } } : age 본래의 값에서 2 증가

#### Document 수정 코드 예제
- people 데이터베이스에서 age가 25 이상인 데이터만 status key의 value를 C로 바꾼다.
    > db.people.updateMany( { age : { $gt : 25 } }, { $set : { status : 'C' } } )
    
    > db.people.updateOne( { age : { $gt : 25) } }, { $set : { status : 'C' } } )

- people 데이터베이스에서 status가 A인 데이터의 age의 값을 3 증가 시킨다.
    > db.people.updateOne( { status : 'A'}, { $inc : { age : 3 } } )
